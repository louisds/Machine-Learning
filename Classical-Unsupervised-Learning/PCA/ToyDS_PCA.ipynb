# PCA on Toy Dataset

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 500

X = np.array(np.random.normal(loc=0,size=2*n))
X = X.reshape(n, 2)
scale = .5
X[:,1] = scale * X[:,1]
# Rotate the dataset
theta = np.pi/6
R = np.array([np.cos(theta), np.sin(theta), -np.sin(theta), 
               np.cos(theta)]).reshape(2,2)
X = X @ R
X[:,0] = X[:,0] - np.mean(X[:,0])
X[:,1] = X[:,1] - np.mean(X[:,1])

plt.figure(figsize=(6,6))
plt.scatter(np.array(X[:,0]), np.array(X[:,1]), alpha=.5)
lim = np.max(np.abs(X))+.1
plt.xlim(-lim,lim)
plt.ylim(-lim,lim)
plt.show()
```

```python
pca = PCA()
pca.fit(X)
principal_vectors = pca.components_
print(pca.explained_variance_)
print(principal_vectors)
```

```python
pc1 = principal_vectors[0]
slope = pc1[1]/pc1[0]

plt.figure(figsize=(6,6))
plt.scatter(np.array(X[:,0]), np.array(X[:,1]), alpha=.5)
plt.plot([lim, -lim], [lim*slope, -lim*slope], color = 'r', linewidth=2)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.show()
```

```python
pcs = pca.fit_transform(X)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(pcs[:,0], pcs[:,1], alpha=.5)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.xlabel("PC1")
plt.ylabel("PC1")
plt.show()
```

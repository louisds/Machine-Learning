# Decision Trees on Iris Dataset

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib.colors import ListedColormap
import random
import matplotlib.pyplot as plt
import sklearn
from sklearn import metrics
%matplotlib inline
```

### Training and Visualizing DTs

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
iris = load_iris()
X = iris.data[:, 2:] # Petal Length and Width
y = iris.target
classes = iris.target_names
classes_dict = dict(zip(range(len(classes)), classes))
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
plot_tree(tree_clf, 
          feature_names=iris.feature_names[2:],
          class_names=iris.target_names,
          rounded=True,
          filled=True)
plt.show()
```

```python
# Probability distribution of classes for a new sample
print(tree_clf.predict_proba([[6, 2]])) 
# Class prediction of a new sample
print(classes_dict[int(tree_clf.predict([[6, 2]]))])
```

```python
fig = plt.figure(figsize = (6, 3))
ax = plt.axes()
colors = ListedColormap(['r','b','g'])

ax.set_title("Classification of Iris Dataset")
ax.set_ylabel("Petal Width")
ax.set_xlabel("Petal Length")
sct = plt.scatter(X[:,0], X[:,1], c=y, cmap=colors)
plt.legend(handles=sct.legend_elements()[0], 
           labels=list(classes))
plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=2)
plt.plot([2.45, 7.5], [1.60, 1.60], "k--", linewidth=2)
plt.text(1.50, 1.0, "Depth=0", fontsize=8)
plt.text(3.2, 1.80, "Depth=1", fontsize=8)
plt.show()
```
